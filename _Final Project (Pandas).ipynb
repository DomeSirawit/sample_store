```python
# TODO 06 - which Segment has the highest profit in 2018
df_2018 = df[df['Order Date'].dt.year == 2018]
df_2018_profit = df_2018.groupby('Segment')['Profit'].sum().reset_index().sort_values('Profit', ascending = False)
df_2018_profit
x = df_2018_profit['Segment'].iloc[0]
y = df_2018_profit['Profit'].iloc[0]
print("The segment that has the most profit in 2018 is (",x,") Segment which has total profit of $", y)
```

```python
# TODO 07 - which top 5 States have the least total sales between 15 April 2019 - 31 December 2019
least_sale_2019 = df[(df['Order Date'] >= '2019-4-15') & (df['Order Date'] <= '2019-12-31')].groupby('State')\
    ['Sales'].sum().sort_values().head().reset_index()
least_sale_2019
```

```python
# TODO 08 - what is the proportion of total sales (%) in West + Central in 2019 e.g. 25% 
# select the all sales from  in 2019 and as the variable then we select only the weste and centreal w+c/all * 100
df_2019 = df[df['Order Date'].dt.year == 2019]
df_2019_sale =  df_2019.groupby('Region')['Sales'].sum().reset_index()

ttl_sal19 = df_2019["Sales"].sum().round()
we_ce_sal19 = df_2019_sale['Sales'].iloc[0].round(2) + df_2019_sale['Sales'].iloc[3].round(2)

sale_perc_wc19 = ((we_ce_sal19 / ttl_sal19)*100).round(1)



print(df_2019_sale.round(2))
print('The total sale in 2019 : $', ttl_sal19)
print('Sales in West and Central region: $', we_ce_sal19)
print('The proprtion of sale of W and C region in 2019: ', sale_perc_wc19, "%" )
```

```python
# TODO 09 - find top 10 popular products in terms of number of orders vs. total sales during 2019-2020

#df[df['col_name'].dt.year.isin([year1, year 2] filter year
#group by Product Name sum Quantity  
# Top product sale in 2019-2020
top_pdt_sl19_20 = df[df["Order Date"].dt.year.isin([2019,2020])]
top10_sl = top_pdt_sl19_20.groupby('Product Name')['Sales'].sum().reset_index().sort_values(by = ['Sales'],ascending=False).round(2).head(10)
top10_sl

                  
```

```python
top_ord10 = top_pdt_sl19_20.groupby('Product Name')['Quantity']\
    .sum().reset_index().sort_values(['Quantity'], ascending = False).head(10)
top_ord10
```

```python
#create column index i to merge
top10_sl['i'] = range(1,len(top10_sl)+1)
top_ord10['i'] = range(1, len(top_ord10) + 1)
```

```python
## merge 2 df
#top_ord10 and top10_sl

top_10_hot = pd.merge(top_ord10, top10_sl, on ='i')
top_10_hot = top_10_hot[['Product Name_x', 'Quantity','Product Name_y', 'Sales']]
top_10_hot.columns = ['Most ordered product', 'Ordered Quantity','Top Sales Product', 'Toal Sales']
top_10_hot
```
